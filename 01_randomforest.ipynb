```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/melb_data.csv')
df.head()
```

```python
#Analyse the data
```

```python
X = df. copy()
X = X.drop(['Price'], axis=1)
X.head()
```

```python
y = df['Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size= 0.8, random_state= 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

```

```python
X_train_full = X_train.copy()
X_test_full = X_test.copy()
y_train_full = y_train.copy()
y_test_full = y_test.copy()
```

```python
X_train.info()
```

```python
cat_columns_original = [col for col in X_train_full if X_train_full[col].dtype == 'object' ]
cat_columns_original, len(cat_columns_original)
```

```python
# Check for null data, get the names of the column with missing data
missing_columns = [col for col in X_train.columns if X_train_full[col].isnull().any()]
missing_columns
```

```python
#Get a list of numerical columns
num_columns = [col for col in X_train_full.columns if X_train_full[col].dtype in ['float64', 'int64']]
len(num_columns)
```

```python
#Get a list of categorical columns
#Cardinality means the number of unique values in a column. Select the columns with relatively low cardinality
cat_columns = [col for col in X_train_full.columns if X_train_full[col].nunique() < 10 and X_train_full[col].dtype == 'object']
cat_columns, len(cat_columns)
```

```python
#Keep the selected columns
my_columns = cat_columns + num_columns
X_train = X_train_full[my_columns]
X_test = X_test_full[my_columns]
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
# Preprocessing for numerical data
num_transformer = SimpleImputer(strategy = 'constant')
```

```python
# Preprocessing for categorical data
cat_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy = 'most_frequent')),
                                  ('onehotcode', OneHotEncoder(handle_unknown='ignore'))])
```

```python
#Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(transformers = [('num', num_transformer, num_columns),
                                                 ('cat', cat_transformer, cat_columns)])
```

```python
#Define the model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
#Create and evaluate the pipeline
from sklearn.metrics import mean_absolute_error

my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', model)])
```

```python
#Fit the model
my_pipeline.fit(X_train, y_train)
```

```python
y_pred = my_pipeline.predict(X_test)
```

```python
score = mean_absolute_error(y_pred, y_test)
score
```
